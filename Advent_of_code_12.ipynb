```python
def rotate(l, n):
    return l[n:] + l[:n]
```

```python
def mult(waypoint, value):
    return [value*i for i in waypoint]

def add(ship,waypoint):
    return [a+b for a,b in zip(waypoint,ship)]
```

```python
def binarytodecimal(string):
    length = len(string) - 1
    value = 0
    for i,j in enumerate(string):
        i = length - i
        pos_value = (2**i)*int(j)
        value = value + pos_value
    return value
```

```python
def operation(mask,value):
    another_output = []
    for i,j in zip(mask,value):
        if i == '1':
            another_output.append('1')
        elif i == '0':
            another_output.append(j)
        elif i == 'X':
            another_output.append('X')
    indices = [i for i,val in enumerate(another_output) if val=='X']
    power = another_output.count('X')
    add = []
    for i in range(0,2**power):
        out = list(bin(i)[2:])
#         print(f'type {type(i)}')
#         decimalToBinary(i)
        if len(out) < power:
            a = [0]*(power - len(out))
            out = [*a,*out]   
        for i,j in zip([*indices],range(power)):
            another_output[i] = str(out[j])
        string = ''.join(another_output)
        add.append(binarytodecimal(string))
        
    return add
```
