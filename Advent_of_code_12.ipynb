```python
initial_pos = [0,0,0,0] # East, South, West, North
direction = 0
```

```python
for i in instructions:
    command = re.findall('\D+',i)[0]
    value = int(re.findall('\d+',i)[0])
    if command == 'R':
        direction = direction + value//90%4
    elif command == 'L':
        direction = direction - value//90%4
    elif command == 'F':
        initial_pos[direction] = initial_pos[direction] + value
    elif command == 'E':
        initial_pos[0] = initial_pos[0] + value
    elif command == 'S':
        initial_pos[1] = initial_pos[1] + value
    elif command == 'W':
        initial_pos[2] = initial_pos[2] + value
    elif command == 'N':
        initial_pos[3] = initial_pos[3] + value
        
    

manhattan_distance = abs(initial_pos[0] - initial_pos[2]) + abs(initial_pos[1] - initial_pos[3])
print(manhattan_distance)
```

```python
def rotate(l, n):
    return l[n:] + l[:n]
```

```python
def mult(waypoint, value):
    return [value*i for i in waypoint]

def add(ship,waypoint):
    return [a+b for a,b in zip(waypoint,ship)]
```

```python
waypoint = [10,0,0,1]
```

```python
initial_pos = [0,0,0,0] # East, South, West, North
direction = 0
```

```python
for i in instructions:
    command = re.findall('\D+',i)[0]
    value = int(re.findall('\d+',i)[0])
    if command == 'R':
        n = value//90
        waypoint = rotate(waypoint,-n)
    elif command == 'L':
        n = value//90
        waypoint = rotate(waypoint,n)
    elif command == 'F':
        initial_pos = add(initial_pos,mult(waypoint,value))
    elif command == 'E':
        waypoint[0] = waypoint[0] + value
    elif command == 'S':
        waypoint[1] = waypoint[1] + value
    elif command == 'W':
        waypoint[2] = waypoint[2] + value
    elif command == 'N':
        waypoint[3] = waypoint[3] + value
    
#     print(f'initial_pos {initial_pos} waypoint {waypoint}')

manhattan_distance = abs(initial_pos[0] - initial_pos[2]) + abs(initial_pos[1] - initial_pos[3])
print(manhattan_distance)
```
