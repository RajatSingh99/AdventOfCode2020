```python
import re
```

```python
filename = '../input_16.txt'
```

```python
with open(filename) as f:
    inp = [line.strip() for line in f]
```

```python
inp
```

```python
re.findall('\d+','type: 34-642 or 648-968')
```

```python
## First Method
# from more_itertools import sliced
# list(sliced(inp,inp.index('')+1))


## Second Method
# output = [[]]

# for x in inp:
#     output[-1].append(x)
#     if x == '':
#         output.append([])
        
## Third Method        
from more_itertools import split_after

# lst = ["a", "b", "c", "a", "d", "c", "e"]
clean_1 = list(split_after(inp, lambda x: x == ""))
# [['a', 'b', 'c'], ['a', 'd', 'c'], ['e']]
```

```python
clean_1
```

```python
clean_1[0][:-1]
```

```python
all_range = []
```

```python
for i in clean_1[0][:-1]:
    temp = re.findall('\d+',i)
    temp = [int(j) for j in temp]
    for k in range(temp[0],temp[1]+1):
        all_range.append(k)
    for k in range(temp[2],temp[3]+1):
        all_range.append(k)
```

```python
all_range = list(set(all_range))
```

```python
all_range
```

```python
output_1 = []
```

```python
for i in clean_1[2][1:]:
    temp = i.split(',')
    temp = [int(j) for j in temp]
    for k in temp:
        if k not in all_range:
            output_1.append(k)
```

```python
sum(output_1)
```

```python
output_1
```

```python
valid_ticket = []
```

```python
my_ticket = [int(k) for k in clean_1[1][1].split(',')]
```

```python
valid_ticket.append(my_ticket)
```

```python
valid_ticket
```

```python
for i in clean_1[2][1:]:
    temp = i.split(',')
    temp = [int(j) for j in temp]
    temp_2 = [i for i in temp if i not in output_1]
    if len(temp) == len(temp_2):
        valid_ticket.append(temp)
```

```python
valid_ticket
```

```python
fields = {}
```

```python
for i in clean_1[0][:-1]:
    key = i.split(':')[0]
    
    temp = re.findall('\d+',i)
    temp = [int(j) for j in temp]
        
    fields[key] = temp
```

```python
fields
```

```python
valid_ticket[1],valid_ticket[2]
```

```python
ticket_individual = []
```

```python
for i in range(0,20):
    temp = []
    for ticket in valid_ticket:
        temp.append(ticket[i])
    ticket_individual.append(temp)
```

```python
ticket_individual
```

```python
for key,value in fields.items():
    all_values = []
    for k in range(value[0],value[1]+1):
        all_values.append(k)
    for k in range(value[2],value[3]+1):
        all_values.append(k)
        
    #print(f'key-{key} value{value} all_value{all_values}')
    #print(f'count of all_values {len(all_values)}')

    for individual in ticket_individual:
        #print(f'individual size {len(individual)}')
        individual.sort()
        check = [i for i in individual if i in all_values]
        #print(check)
        #print(f'check size {len(check)}')
        if check == individual:
            print(f'{key} is at index {ticket_individual.index(individual)}')
```
