```python
def check_field(temp):
    final_value = True
    eye_colour = ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
    hcl_pattern = '^#[0-9a-f]{6}'
    hgt_pattern = '[0-9]+(cm|in)'
    for i in temp:
        field = i.split(':')[0]
        value = i.split(':')[1]
        if field == 'byr':
            if (len(value) == 4) and (int(value) >= 1920) and (int(value) <= 2002):
                continue
            else:
                final_value = False
                break

        elif field == 'iyr':
            if (len(value) == 4) and (int(value) >= 2010) and (int(value) <= 2020):
                continue
            else:
                final_value = False
                break

        elif field == 'eyr':
            if (len(value) == 4) and (int(value) >= 2020) and (int(value) <= 2030):
                continue
            else:
                final_value = False
                break

        elif field == 'hgt':
            if re.match(hgt_pattern,value):
                height = int(re.findall('\d+',value)[0])
                if re.findall('\D+',value)[0] == 'cm':
                    if (height >= 150) and (height <= 193):
                        continue
                    else:
                        final_value = False
                        break
                elif re.findall('\D+',value)[0] == 'in':
                    if (height >= 59) and (height <= 76):
                        continue
                    else:
                        final_value = False
                        break
                else:
                    final_value = False
                    break
            else:
                final_value = False
                break
                
        elif field == 'hcl':
            if re.match(hcl_pattern,value):
                continue
            else:
                final_value = False
                break

        elif field == 'ecl':
            if (value in eye_colour):
                continue
            else:
                final_value = False
                break

        elif field == 'pid':
            if (len(value) == 9):
                continue
            else:
                final_value = False
                break

    return final_value
```

```python
def binary_value(string):
    length = len(string) - 1
    value = 0
    for i,j in enumerate(string):
        i = length - i
        pos_value = (2**i)*int(j)
        value = value + pos_value
    return value
```
