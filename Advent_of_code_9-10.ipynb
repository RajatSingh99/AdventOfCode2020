```python
while (start + 3) <= max_output:
    if (start+1 in input_volts) or (start+1 == max_output):
        diff_1 = diff_1 + 1
        start = start+1
    elif (start+2 in input_volts) or (start+2 == max_output):
        diff_2 = diff_2 + 1
        start = start+2
    elif (start+3 in input_volts) or (start+3 == max_output):
        diff_3 = diff_3 + 1
        start = start+3
```

```python
def addition(volt):
    if (volt != 0) and (volt not in volts.keys()):
        return 0
    elif volts[volt] > 0:
        return volts[volt]
    else:
        volts[volt] = addition(volt-1) + addition(volt-2) + addition(volt-3)
        return volts[volt]
    
```
